# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
import numpy as np

def bernoulli_pdf(x, p):
  """
  Calculates the probability density function (PDF) of a Bernoulli distribution.

  Args:
    x: The value of the random variable (0 or 1).
    p: The probability of success (probability of x=1).

  Returns:
    The probability density function value.
  """

  if x not in [0, 1]:
    raise ValueError("x must be either 0 or 1")

  if not (0 <= p <= 1):
    raise ValueError("p must be between 0 and 1")

  if x == 1:
    return p
  else:
    return 1 - p

# Example usage
p = 0.7  # Probability of success
x_values = [0, 1]

for x in x_values:
  pdf_value = bernoulli_pdf(x, p)
  print(f"P(X={x}) = {pdf_value}")
```

# 9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
  """
  Calculates Z-scores for a given dataset.

  Args:
    data: A list or NumPy array containing the data.

  Returns:
    A NumPy array of Z-scores.
  """
  mean = np.mean(data)
  std_dev = np.std(data)
  z_scores = (data - mean) / std_dev
  return z_scores

# Example usage
data = [12, 15, 18, 21, 24, 27, 30, 33, 36, 39]
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.figure(figsize=(8, 5))
plt.plot(z_scores, 'bo-', label='Z-scores')
plt.axhline(y=0, color='r', linestyle='--', label='Mean')
plt.xlabel('Data Points')
plt.ylabel('Z-score')
plt.title('Z-scores of the Dataset')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameters
population_size = 10000
sample_size = 30
num_samples = 1000

# Generate population data from an exponential distribution
population = expon.rvs(scale=1, size=population_size)

# Initialize a list to store sample means
sample_means = []

# Draw samples and calculate their means
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b', label='Sample Means')

# Calculate and plot the theoretical normal distribution
population_mean = np.mean(population)
population_std = np.std(population)
standard_error = population_std / np.sqrt(sample_size)

x = np.linspace(population_mean - 3*standard_error, population_mean + 3*standard_error, 100)
plt.plot(x, 1/(standard_error * np.sqrt(2 * np.pi)) * np.exp( - (x - population_mean)**2 / (2 * standard_error**2)),
         'r-', linewidth=2, label='Normal Distribution')

# Customize the plot
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Central Limit Theorem: Distribution of Sample Means')
plt.legend()
plt.grid(True)
plt.show()
```

# 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
  """
  Calculates the Z-score for a given data point.

  Args:
    data_point: The data point for which to calculate the Z-score.
    mean: The mean of the population.
    std_dev: The standard deviation of the population.

  Returns:
    The calculated Z-score.
  """
  z_score = (data_point - mean) / std_dev
  return z_score

def plot_standard_normal_with_z_score(z_score):
  """
  Plots the standard normal distribution and highlights the given Z-score.

  Args:
    z_score: The Z-score to highlight on the plot.
  """
  x = np.linspace(-3, 3, 100)
  pdf = norm.pdf(x)

  plt.figure(figsize=(8, 5))
  plt.plot(x, pdf, color='blue', label='Standard Normal PDF')
  plt.fill_between(x[x <= z_score], pdf[x <= z_score], color='red', alpha=0.25)
  plt.axvline(x=z_score, color='red', linestyle='--', label=f'Z-score: {z_score:.2f}')
  plt.xlabel('Z-score')
  plt.ylabel('Probability Density')
  plt.title('Standard Normal Distribution with Z-score')
  plt.legend()
  plt.grid(True)
  plt.show()

# Example usage
data_point = 75
mean = 70
std_dev = 5

z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score: {z_score}")

plot_standard_normal_with_z_score(z_score)
```

# 15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
  """
  Performs a one-sample Z-test.

  Args:
    sample_mean: The mean of the sample.
    population_mean: The hypothesized population mean.
    population_std: The population standard deviation.
    sample_size: The size of the sample.
    alpha: Significance level (default: 0.05).

  Returns:
    z_statistic: The calculated Z-statistic.
    p_value: The p-value.
    reject_null: True if the null hypothesis is rejected, False otherwise.
  """

  z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
  p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed test
  reject_null = p_value < alpha

  return z_statistic, p_value, reject_null

# Example usage
sample_mean = 102
population_mean = 100
population_std = 15
sample_size = 50

z_stat, p_value, reject_null = z_test(sample_mean, population_mean, population_std, sample_size)

print(f"Z-statistic: {z_stat}")
print(f"p-value: {p_value}")
print(f"Reject null hypothesis: {reject_null}")
```

# 16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence=0.95):
  """
  Calculates the confidence interval for a given dataset.

  Args:
    data: A list or NumPy array containing the data.
    confidence: Confidence level (default: 0.95).

  Returns:
    A tuple containing the lower and upper bounds of the confidence interval.
  """

  n = len(data)
  mean = np.mean(data)
  std_err = stats.sem(data)  # Standard error of the mean

  # Calculate the critical value (t-score)
  t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

  # Calculate the margin of error
  margin_of_error = t_critical * std_err

  # Calculate the lower and upper bounds of the confidence interval
  lower_bound = mean - margin_of_error
  upper_bound = mean + margin_of_error

  return lower_bound, upper_bound

# Example usage
data = [25, 28, 22, 27, 23, 25, 26, 24, 29, 26]
confidence_level = 0.95  # 95% confidence interval

lower_bound, upper_bound = calculate_confidence_interval(data, confidence_level)

print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats

def generate_and_analyze_normal_data(sample_size, confidence_level=0.95):
  """
  Generates data from a standard normal distribution, calculates and interprets
  the confidence interval for its mean.

  Args:
    sample_size: The size of the sample to generate.
    confidence_level: The desired confidence level for the interval.

  Returns:
    A tuple containing:
      - sample_mean: The mean of the generated sample.
      - confidence_interval: A tuple containing the lower and upper bounds.
      - interpretation: A string describing the confidence interval.
  """

  # Generate data from a standard normal distribution
  data = np.random.normal(loc=0, scale=1, size=sample_size)

  # Calculate sample mean and standard error
  sample_mean = np.mean(data)
  standard_error = np.std(data, ddof=1) / np.sqrt(sample_size)

  # Calculate critical value (z-score)
  z_critical = stats.norm.ppf((1 + confidence_level) / 2)

  # Calculate margin of error
  margin_of_error = z_critical * standard_error

  # Calculate confidence interval
  lower_bound = sample_mean - margin_of_error
  upper_bound = sample_mean + margin_of_error
  confidence_interval = (lower_bound, upper_bound)

  # Interpretation
  interpretation = f"With {confidence_level:.0%} confidence, the true population mean " \
                  f"lies between {lower_bound:.2f} and {upper_bound:.2f}."

  return sample_mean, confidence_interval, interpretation

# Example usage
sample_size = 50
confidence_level = 0.95

sample_mean, confidence_interval, interpretation = generate_and_analyze_normal_data(sample_size, confidence_level)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: {confidence_interval}")
print(interpretation)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
  """
  Compares the standard deviations of two datasets and visualizes the difference.

  Args:
    data1: The first dataset.
    data2: The second dataset.

  Returns:
    None
  """

  std_dev1 = np.std(data1)
  std_dev2 = np.std(data2)

  print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
  print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

  # Visualize the difference (optional)
  plt.figure(figsize=(8, 5))
  plt.bar(['Dataset 1', 'Dataset 2'], [std_dev1, std_dev2])
  plt.ylabel('Standard Deviation')
  plt.title('Comparison of Standard Deviations')
  plt.show()

# Example usage
data1 = [10, 12, 15, 18, 20, 22, 25]
data2 = [5, 8, 11, 14, 17, 20, 23]

compare_standard_deviations(data1, data2)
```

```python
import numpy as np

# Parameters
mu = 0  # Mean of the normal distribution
sigma = 1  # Standard deviation of the normal distribution
sample_size = 1000  # Sample size

# Generate data from normal distribution
data = np.random.normal(mu, sigma, sample_size)

# Calculate range
data_range = np.max(data) - np.min(data)

# Calculate IQR
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1

# Print results
print(f"Range: {data_range:.4f}")
print(f"Interquartile Range (IQR): {iqr:.4f}")
```

# 23. Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
  """
  Performs Z-score normalization on the given dataset.

  Args:
    data: A NumPy array or list containing the data.

  Returns:
    A NumPy array containing the Z-score normalized data.
  """
  mean = np.mean(data)
  std_dev = np.std(data)
  z_scores = (data - mean) / std_dev
  return z_scores

# Sample dataset
data = np.array([1, 5, 3, 9, 2, 7, 4])

# Perform Z-score normalization
z_scores = z_score_normalization(data)

# Visualize the transformation
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=10, alpha=0.5, label='Original Data')
plt.xlabel('Original Values')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=10, alpha=0.5, label='Normalized Data')
plt.xlabel('Z-scores')
plt.ylabel('Frequency')
plt.legend()

plt.suptitle('Z-Score Normalization')
plt.show()
```
